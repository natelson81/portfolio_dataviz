```python
import pandas as pd
import plotly.graph_objects as go
from plotly.colors import n_colors
#import numpy as np
from IPython.display import Image
```

```python
df1 = pd.read_csv('csv/vacinacao_ufs.csv', sep=";")
df2 = pd.read_csv('csv/vacinacao_capitais.csv', sep=";")
```

### TABELAS POR ESTADO

```python
ufs = df1.iloc[:-1,0:5].dropna()

colors = n_colors('rgb(153,215,235)', 'rgb(242,252,254)',  len(ufs), colortype='rgb')
#np.random.seed(1)

b = 'white'
c = '#61aac3'

fig = go.Figure(data=[go.Table(columnwidth = [25,20,20,20,20],
                               header=dict(values=ufs.columns,
                                           align='center',
                                           fill_color='#61aac3', 
                                           font=dict(color='white', size=12)),
                         
                 cells=dict(values=[ufs.iloc[:,0],            
                                    ufs.iloc[:,1], 
                                    ufs.iloc[:,2], 
                                    ufs.iloc[:,3],
                                    ufs.iloc[:,4]
                                   ],
                            align='center',
                            font=dict(color='black', size=12),
                            fill=dict(color=[c, b, b, b, colors])
                           )),
                     ]
               )
fig.update_layout(title='% de Totalmente Imunizados por Estado',
                  autosize=True,
                  width=750,
                  height=490,
                  margin=dict(l=20,r=20,b=30,t=80,pad=3),
                  #paper_bgcolor="LightSteelBlue",
                 )

#fig.show()

# Mostrando o grfico em formato esttico. Github no renderiza formato dinmico da biblioteca Plotly.
img_bytes = fig.to_image(format="png", width=800, height=350, scale=2)
Image(img_bytes)
```

```python
ufs_doses = df1.loc[0:26,["Estado", "Doses aplicadas", "Doses distribudas", "% doses usadas"]].sort_values("% doses usadas", ascending=False)

colors = n_colors('rgb(153,215,235)', 'rgb(242,252,254)',  len(ufs_doses), colortype='rgb')
#np.random.seed(1)

b = 'white'
c = '#61aac3'

fig = go.Figure(data=[go.Table(columnwidth = [15,20,20,20],
                               header=dict(values=ufs_doses.columns,
                                           align='center',
                                           fill_color='#61aac3', 
                                           font=dict(color='white', size=12)),
                         
                 cells=dict(values=[ufs_doses["Estado"],            
                                    ufs_doses["Doses aplicadas"],
                                    ufs_doses["Doses distribudas"],
                                    ufs_doses["% doses usadas"]
                                   ],
                            align='center',
                            font=dict(color='black', size=12),
                            fill=dict(color=[c, b, b, colors])
                           )),
                     ]
               )
fig.update_layout(title='% de Doses Usadas por Estados',
                  autosize=True,
                  width=750,
                  height=490,
                  margin=dict(l=20,r=20,b=30,t=80,pad=3),
                  #paper_bgcolor="LightSteelBlue",
                 )

#fig.show()

# Mostrando o grfico em formato esttico. Github no renderiza formato dinmico da biblioteca Plotly.
img_bytes = fig.to_image(format="png", width=800, height=350, scale=2)
Image(img_bytes)
```
