```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.colors import n_colors
import numpy as np
from IPython.display import Image
```

```python
tab_leitos = pd.read_csv('csv/ocup_leitos.csv', sep=';')

tab_leitos_7d = tab_leitos.sort_values(['Data', '% de ocupação de UTI'], ascending=False).head(189)
```

### TABELA DE OCUPAÇÃO DE LEITOS DE UTI:

```python
colors = n_colors('rgb(255,255,255)', 'rgb(223,132,0))',  len(tab_leitos_7d), colortype='rgb')
np.random.seed(1)

b = 'white'
c = '#61aac3'
     
fig = go.Figure(data=[go.Table(columnwidth = [10,12,10,10],
                               header=dict(values=tab_leitos_7d.columns,
                                           align='center',
                                           fill_color='#61aac3', 
                                           font=dict(color='white', size=12)),
                         
                 cells=dict(values=[tab_leitos_7d['Regiao'],            
                                    tab_leitos_7d['Estado'], 
                                    tab_leitos_7d['Data'], 
                                    tab_leitos_7d['% de ocupação de UTI'],
                                   ],
                            align='center',
                            font=dict(color='black', size=12),
                            fill=dict(color=[c, b, b, np.array(colors)[tab_leitos_7d['% de ocupação de UTI'].values]]),
                           )),
                     ]
               )
fig.update_layout(title='Taxa de Ocupação de Leitos de UTI por Região e Estados (Últimos 7 dias)',
                  autosize=False,
                  width=750,
                  height=490,
                  margin=dict(l=20,r=20,b=30,t=80,pad=3),
                  #paper_bgcolor="LightSteelBlue",
                 )

#fig.show()

# Mostrando o gráfico em formato estático. Github não renderiza formato dinâmico da biblioteca Plotly.
img_bytes = fig.to_image(format="png", width=800, height=550, scale=2)
Image(img_bytes)
```
