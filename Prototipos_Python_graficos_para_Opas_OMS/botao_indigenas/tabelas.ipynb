```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.colors import n_colors
import numpy as np
from IPython.display import Image
```

```python
dseis1 = pd.read_csv('csv/dseis1.csv', sep=';')
dseis1 = dseis1.dropna()
dseis2 = pd.read_csv('csv/dseis2.csv', sep=';')
dseis2 = dseis2.dropna()
```

```python
colors = n_colors('rgb(255,255,255)', 'rgb(223,132,0))',  len(dseis1), colortype='rgb')
np.random.seed(1)

b = 'white'
c = '#61aac3'

fig = go.Figure(data=[go.Table(columnwidth = [20,15,15,20,20,25,25,25,10],
                               header=dict(values=dseis1.columns,
                                           align='center',
                                           fill_color='#61aac3', 
                                           font=dict(color='white', size=10)
                                          ),

                 cells=dict(values=[dseis1.iloc[:,0],
                                   dseis1.iloc[:,1],
                                   dseis1.iloc[:,2],
                                   dseis1.iloc[:,3],
                                   dseis1.iloc[:,4],
                                   dseis1.iloc[:,5],
                                   dseis1.iloc[:,6],
                                   dseis1.iloc[:,7],
                                   dseis1.iloc[:,8],
                                   ],
                            align='center',
                            font=dict(color='black', size=10),
                            #fill=dict(color=[c, b, b, b, b, b,b,b,b]),
                           )),
                     ]
               )
fig.update_layout(title='TABELA 01 (CASOS): Resumo da situao nos Distritos Sanitrios Especiais Indgenas (DSEI)',
                  autosize=True,
                  width=1050,
                  height=490,
                  margin=dict(l=20,r=20,b=100,t=80,pad=3),
                  #paper_bgcolor="LightSteelBlue",
                 )

#Notas da Tabela:
fig.add_annotation(x=0.01, y=-0.05, showarrow=False, text="NOTAS:", font=dict(color='black', size=9),),
fig.add_annotation(x=0.01, y=-0.10, showarrow=False, text="1. Atualizao semanal;", font=dict(color='black', size=9),),
fig.add_annotation(x=0.01, y=-0.15, showarrow=False, text="2. Por data de incio de sintomas;", font=dict(color='black', size=9)),
fig.add_annotation(x=0.01, y=-0.20, showarrow=False, text="3. Taxa por 100.000 habitantes;", font=dict(color='black', size=9))
fig.add_annotation(x=0.01, y=-0.25, showarrow=False, text="4. Nmero de Reproduo. Apenas DSEI com casos reportados nos ltimos 30 dias foram analisados;", font=dict(color='black', size=9))
fig.add_annotation(x=0.01, y=-0.30, showarrow=False, text="*. SE: Semana Epidemiolgica.", font=dict(color='black', size=9))

fig.add_annotation(x=1, y=-0.05, showarrow=False, text="Fonte: Secretaria Especial de Sade Indgena", font=dict(color='black', size=9),)

#fig.show()

# Mostrando o grfico em formato esttico. Github no renderiza formato dinmico da biblioteca Plotly.
img_bytes = fig.to_image(format="png", width=800, height=650, scale=2)
Image(img_bytes)
```
