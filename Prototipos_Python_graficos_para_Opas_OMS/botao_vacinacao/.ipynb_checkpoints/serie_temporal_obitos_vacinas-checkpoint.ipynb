```python
import pandas as pd
import plotly
from plotly.colors import n_colors
import plotly.graph_objs as go
import plotly.express as px
import numpy as np
from IPython.display import Image
```

#### GRÁFICO DE 2 EIXOS Y - ÓBITOS POR COVID-19 X VACINAÇÃO - DADOS RANDÔMICOS 

```python
colors = {'A':'#d46c2c',
          'B':'#1490a8','C':'#1783df', 'D':'#407f36'}

semana_epidemiologica = []

i = 1
while len(semana_epidemiologica) < 50:
    semana_epidemiologica.append(i)
    i += 1
    
x = semana_epidemiologica

trace1 = go.Bar(
    x=semana_epidemiologica,
    y=np.random.randint(low=1, high=500, size=len(semana_epidemiologica)),
    name='Óbito por SRAG confirmado para Covid-19', 
    marker_color=colors['A'],
    opacity=0.6,
    )

trace2 = go.Bar(
    x=semana_epidemiologica,
    y=np.random.randint(low=1, high=800, size=len(semana_epidemiologica)),
    name='Óbito por SRAG não especificado',
    marker_color=colors['B'],
    opacity=0.6,
    )

trace3 = go.Scatter(
    x=semana_epidemiologica,
    y=np.random.randint(low=200, high=500, size=len(semana_epidemiologica)),
    name='Vacinação 1ª dose',
    marker_color=colors['C'], 
    yaxis='y2',
    )

trace4 = go.Scatter(
    x=semana_epidemiologica,
    y=np.random.randint(low=200, high=500, size=len(semana_epidemiologica)),
    name='Vacinação 2ª dose',
    marker_color=colors['D'],
    )

data=[trace1, trace2, trace3, trace4]

layout= go.Layout(
    barmode= 'stack',
    bargap=0.0,
    title='Óbitos por COVID-19 e doses de vacina aplicada por semana epidemiológica',
    xaxis=dict(title='Semana Epidemiológica'),
    yaxis=dict(title='Óbitos'),
    yaxis2=dict(title='Doses aplicadas(/100hab.)',
                overlaying='y',
                side='right')
    )
fig= go.Figure(data=data, layout=layout)
fig.update_layout(legend=dict(orientation="h", yanchor="bottom", y=-0.7, xanchor="right", x=1,))
fig.update_layout(yaxis = dict(tickfont = dict(size=10)))

#fig.show()

# Mostrando o gráfico em formato estático. Github não renderiza formato dinâmico da biblioteca Plotly.
img_bytes = fig.to_image(format="png", width=800, height=350, scale=2)
Image(img_bytes)
```
