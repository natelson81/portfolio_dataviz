```python
import pandas as pd
import plotly
from plotly.colors import n_colors
import plotly.graph_objs as go
import plotly.express as px
import numpy as np
from IPython.display import Image
```

#### Distribuição das sequências de variantes circulantes no Brasil

```python
df_copias = pd.read_csv("csv/variantes_copias.csv", sep=";")
df = df_copias

fig = px.bar(df, 
             x="Data", 
             y="Copias", 
             color="Variante", 
             title="Total de Cópias por Variante", 
             barmode='stack',
             text='Copias', 
             width=800, 
             height=400)

fig.update_layout(legend=dict(orientation="h", yanchor="top", y=1.15, xanchor="right", x=1,))
fig.update_layout(yaxis = dict(tickfont = dict(size=10)))
fig.update_layout(title='Distribuição das sequências de variantes circulantes no Brasil segundo data de coleta da amostra',
                  autosize=False,
                  title_font_size=14)

#fig.show()

# Mostrando o gráfico em formato estático. Github não renderiza formato dinâmico da biblioteca Plotly.
img_bytes = fig.to_image(format="png", width=800, height=350, scale=2)
Image(img_bytes)
```

```python
df_delta = df_copias[df_copias['Variante'] == 'Delta']

fig = px.bar(df_delta, 
             x="Data", 
             y="Copias", 
             #color="Variante", 
             title="Total de Cópias da Variante Delta no Brasil", 
             barmode='stack',
             text='Copias', 
             width=800, 
             height=400)

fig.update_layout(legend=dict(orientation="h", yanchor="top", y=1.15, xanchor="right", x=1,))
fig.update_layout(yaxis = dict(tickfont = dict(size=10)))

#fig.show()

# Mostrando o gráfico em formato estático. Github não renderiza formato dinâmico da biblioteca Plotly.
img_bytes = fig.to_image(format="png", width=800, height=350, scale=2)
Image(img_bytes)
```

```python
delta_gisaid = pd.read_csv('csv/casos_delta_GISAID.csv', sep=';')
delta_ms = pd.read_csv('csv/casos_delta_MS.csv', sep=';')
```

```python
colors = n_colors('rgb(153,215,235)', 'rgb(242,252,254)',  len(delta_gisaid), colortype='rgb')
#np.random.seed(1)

b = 'white'
c = '#61aac3'

fig = go.Figure(data=[go.Table(columnwidth = [10,20,20],
                               header=dict(values=delta_gisaid.columns,
                                           align='center',
                                           fill_color='#61aac3', 
                                           font=dict(color='white', size=12)),
                         
                 cells=dict(values=[delta_gisaid.iloc[:,0],            
                                    delta_gisaid.iloc[:,1], 
                                    delta_gisaid.iloc[:,2], 
                                   ],
                            align='center',
                            font=dict(color='black', size=11),
                            fill=dict(color=[c, b, b])
                           )),
                     ]
               )
fig.update_layout(title='Casos notificados pelo ponto focal do RSI sequências postadas no GISAID',
                  autosize=False,
                  title_font_size=11,
                  width=450,
                  height=490,
                  margin=dict(l=20,r=20,b=30,t=80,pad=3),
                  #paper_bgcolor="LightSteelBlue",
                 )

#fig.show()

# Mostrando o gráfico em formato estático. Github não renderiza formato dinâmico da biblioteca Plotly.
img_bytes = fig.to_image(format="png", width=800, height=350, scale=2)
Image(img_bytes)
```
